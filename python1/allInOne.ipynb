```python
import pickle as pkl
from pyKriging.krige import kriging
```

```python
df=pd.read_excel('sobol结果1203inp.xlsx',header = None)
x = np.array(df)

dff=pd.read_excel('sobol结果1203outp.xlsx',header = None)
y = np.array(dff)
y=y.flatten()
```

```python
k = kriging(x[::50],y[::50])
k.train()
with open('kriging_saved.pkl','wb') as f:
    pkl.dump(k,f)
# with open('kriging_saved.pkl','rb') as f:
#     k_saved = pkl.load(f)
k.predict([0.488586,	0.090759,	0.484436067,	0.151306133,	0.399230933
])
```

```python
from SALib.analyze import sobol
from SALib.sample import saltelli
from SALib.test_functions import Ishigami
from SALib.util import read_param_file
```

```python
problem = {
    'num_vars': 5,
    'names': ['A14', 'A23','Tup','Tmid','Tdown'],
    'bounds': [[0.095, 0.1],
               [0.09, 0.095],
               [0,150],
               [0,150],
               [0,150]]
}
#%%
param_values = saltelli.sample(problem, 5000,calc_second_order=False)
```

### 4.2 读取灵敏度分析相关有限元计算结果

```python
columns=['A14','A23','Tup','Tmid','Tdown','23x','23y','23z']
df=pd.read_excel('灵敏度分析35000.xlsx',header = None, names=columns)
Z = df['23z'].values

Si = sobol.analyze(problem, Z, calc_second_order=False, print_to_console=True)
print()
```

```python
#%% Z灵敏度
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

#设置x轴取值
xedges = np.array([10,20,30,40,50,60])
#设置y轴取值
yedges = np.array([10,20,30])
#设置X,Y对应点的值。即原始数据。
hist =np.array( [[0.284421,0.285008],
                 [0.372754,0.373051],
                 [0.333049,0.333180],
                 [0.000238,0.000873],
                 [0.006613,0.006513]])

#生成图表对象。
fig = plt.figure()
#生成子图对象，类型为3d
ax = fig.add_subplot(111,projection='3d')

#设置作图点的坐标
xpos, ypos = np.meshgrid(xedges[:-1]-2.5 , yedges[:-1]-2.5 )
xpos = xpos.flatten('F')
ypos = ypos.flatten('F')
zpos = np.zeros_like(xpos)

#设置柱形图大小
dx =5 * np.ones_like(zpos)
dy = dx.copy()
dz = hist.flatten()

#设置坐标轴标签
ax.set_xlabel('QOI')
ax.set_ylabel('S1ST')
ax.set_zlabel('sensitivity')

# x, y, z: array - like
# The coordinates of the anchor point of the bars.
# dx, dy, dz: scalar or array - like
# The width, depth, and height of the bars, respectively.
# minx = np.min(x)
# maxx = np.max(x + dx)
# miny = np.min(y)
# maxy = np.max(y + dy)
# minz = np.min(z)
# maxz = np.max(z + dz)
ax.bar3d(xpos, ypos, zpos, dx, dy, dz,color='b',zsort='average')
ax.set_zlim3d(0, 0.6)
plt.show()
```
